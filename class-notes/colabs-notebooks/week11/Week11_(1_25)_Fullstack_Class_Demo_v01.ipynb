#Manually deploy react app
----------------------------------

1. create a public repository or use an existing repoitory
2. create "gh-pages" branch or is exist delete all files from "gh-pages" branch
3. build react project
4. upload react build files to gh-pages branch
5. delete repository readme.md file if exist
   note: you should now be able to browse to your react app at the repo github pages: setting/pages
6. re-build your react app
7. delete repo "gh-pages" branch files
6. re-upload build files to repo "gh-pages"


- note: github will enable github pages for this repo and url
- preview in repository settings/pages
